```python
import pandas as pd

df = pd.read_csv("../data/posts.csv")
print(df.shape)
print(df.head())
```

```python
print("Missing values:\n", df.isnull().sum())
print("\nEmpty text posts:", (df['text'] == '').sum())
print("\nScore range:", df['score'].min(), "to", df['score'].max())
print("\nAvg comments per post:", df['num_comments'].mean().round(2))
```

```python
# fill missing text with the title so we don't lose those 90 posts
df['text'] = df['text'].fillna(df['title'])

# convert unix timestamp to readable date
df['date'] = pd.to_datetime(df['created_utc'], unit='s')

# check result
print(df[['title', 'text', 'date']].head())
print("\nMissing text after fix:", df['text'].isnull().sum())
```

```python
df.to_csv("../data/posts_cleaned.csv", index=False)
print("Saved cleaned data")
```
